```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
df
```

```python
df_train = df.iloc[:int(len(df) * 0.7)]
df_test = df.iloc[int(len(df) * 0.7):]
df_train
```

```python
df_test
```

```python
features = ["PetalLengthCm","PetalWidthCm"]

sns.scatterplot(data=df_train,x=features[0],y=features[1],hue="Species")

plt.title("Training Data for KNN of Iris Dataset")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.show()
```

```python
features = ["PetalLengthCm","PetalWidthCm"]

sns.scatterplot(data=df_test,x=features[0],y=features[1],hue="Species")

plt.title("Testing Data for KNN of Iris Dataset")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.show()
```
