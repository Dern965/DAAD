## Iris

Leemos el dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn import datasets
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
df = pd.read_csv("Iris.csv")
df
```

Dividimos el dataset en 70% para entrenar y 30% para probar

```python
x = df.drop("Species", axis=1)
y = df["Species"]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)

df_train = pd.concat([x_train,y_train],axis=1)
df_test = pd.concat([x_test,y_test],axis=1)
```

Gráficamos los datos de entrenamiento y prueba

Calculamos las distancias euclidianas

```python
def dist_euclidiana(p1,p2):
    return np.sqrt(np.sum(p1-p2)**2)
```

### KNN

```python
def knn(train_x,train_y,point,k):
    distance=[]
    for i in range(len(train_x)):
        dist=dist_euclidiana(train_x[i],point)
        distance.append(dist)
    distances=np.array(distance)
    near_neighbors = np.argsort(distances)[:k]
    near_clases = train_y[near_neighbors]
    unique_clases, count_clases = np.unique(near_clases,return_counts=True)
    
    return unique_clases[np.argmax(count_clases)]
```

#### Test

```python
k_values = [1,3,5]

predictions_dict = {}

for k in k_values:
    predictions = []
    for i in range(len(x_test)):
        test_point = x_test.iloc[i].values
        prediction = knn(x_train.values,y_train.values,test_point,k)
        predictions.append(prediction)

    predictions_dict[k] = predictions

for k in k_values:
    print(f"Predictions for k={k}")
    for i in range(len(x_test)):
        print(f"Prediction for the point {i+1}: Prediction = {predictions[i]}")
```

### Matriz de confusion

```python
confusion = confusion_matrix(y_test,predictions)
print("Confusion Matrix")
print(confusion)
```

#### Calculamos metricas

```python
accuracy = accuracy_score(y_test, predictions)
precision = precision_score(y_test, predictions, average='weighted')
recall = recall_score(y_test, predictions, average='weighted')
f1 = f1_score(y_test, predictions, average='weighted')

print(f"Precisión: {accuracy}")
print(f"Precisión: {precision}")
print(f"Exhaustividad: {recall}")
print(f"Puntuación F1: {f1}")
```
