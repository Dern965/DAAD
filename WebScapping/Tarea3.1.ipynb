```python
from bs4 import BeautifulSoup
import requests
import bs4
import pandas as pd
```

Checamos las versiones

Definimos la url de la pagina a la que vamos hacer solicitud

```python
url = "https://www.amazon.com.mx/s?k=libros"
response = requests.get(url)

if response.status_code == 200:
    html = response.text
    print("Conexion exitosa:\n", html)
else:
    print(f"Error en la conexion: {response.status_code}")
```

Hacemos una sopa xd

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
type(soup)
```

```python
items_title = soup.find_all("h2", class_="a-size-mini a-spacing-none a-color-base s-line-clamp-4")
items_title
```

```python
titles = []
for item in items_title:
    title = item.find('a').text.strip()
    titles.append(title)
titles
```

```python
prices = []
items_price = soup.find_all("span", class_="a-price-whole")
for item in items_price:
    price = item.text.strip()
    price = price.replace(".", "")
    price = int(price)
    prices.append(price)
prices
```

```python
pennies = []
items_penny = soup.find_all("span", class_="a-price-fraction")
for item in items_penny:
    penny = item.text.strip()
    penny = penny.replace(".", "")
    penny = int(penny)
    pennies.append(penny)
pennies
```

```python
del pennies[3]
```

```python
ratings = []
items_rating = soup.find_all("span", class_="a-icon-alt")
for item in items_rating:
    rating = item.text.strip()
    ratings.append(rating)
ratings
```

```python
ratings = ratings[:52]
```

```python
prices_f = []
for int, cent in zip(prices, pennies):
    prices2 = f"{int}.{cent}"
    prices_f.append(prices2)
prices_f
```

```python
df = pd.DataFrame({'Titulo': titles, 'Calificacion': ratings, 'Precio': prices_f})
```
